## Cross Validation
#### 
Análise da validação cruzada no contexto de Machine Learning serve como uma ferramenta essencial para avaliar o desempenho de um modelo de maneira mais precisa e confiável, ajudando a mitigar problemas como o overfitting e a variabilidade dos resultados

A validação cruzada permite avaliar o modelo em várias divisões, reduzindo a chance de um modelo enviesado ou ajustado demais a um subconjunto específico.

#### 1. Método Hold Out
Isso é simplesmente dividir os dados em conjunto de treinamento e teste. A porcentagem de dados de treinamento é maior que os dados de teste. Usa o conjunto de treinamento para treinar o modelo e o conjunto de teste restante para estimativa de erro. Pode gerar resultados variáveis e não é ideal para datasets pequenos.

#### 2. Validação cruzada K-fold:
O  conjunto de dados inteiro n é dividido em k partes com n/k =p e então esse p será tomado como dados de teste em cada iteração e o próximo p na próxima iteração e assim por diante até k iterações. Os conjuntos serão diferentes para diferentes dobra e cadaa dado será considerado uma vez no conjunto de teste e k-1 vezes no conjunto de treinamento, aumentando a eficácia deste método . A  precisão final será a média de todas essas 5 precisões. Além disso, seremos capazes de obter a precisão mínima e máxima deste modelo específico.

#### 3. Validação cruzada de deixar um de fora (LOOCV):
Para todos os pontos de dados, um dado é deixado como dado de teste e o restante como dado de treinamento. Então, para n pontos de dados, temos que executar n iterações para cobrir cada ponto de dados. (Não tão eficiente)

Apesar de reduzir o viés, é computacionalmente caro e pode levar ao overfitting.

#### 4. Validação cruzada estratificada:
Nisso, a amostra aleatória preenchida no conjunto de dados de treinamento e teste é tal que o número de instâncias de cada classe em cada iteração de divisão de dados de treinamento e teste é considerado em boa proporção de sim e não, 0 e 1, ou altos e baixos, para que o modelo forneça boa precisão

Ideal para datasets desbalanceados.

#### 5. Divisões de testes aleatórios repetidos :
Aa divisão aleatória do conjunto de dados é feita em conjunto de trinamentom e teste e, em seguida, o processo adicional de divisão e medição de desempenho é repetido pelo número de vezes especificao s. A validação cruzada é realizada

É eficiente para grandes datasets, mas não garante que todos os dados sejam utilizados em cada iteração.



```python
import pandas as pd 
```

```python
data = pd.read_csv("datasets/data.csv")
data
```

```python
data.isnull().sum()
```

```python
data1 = data.drop(['Unnamed: 32'], axis='columns')
data1.isnull().sum()
```

```python
# Dividindo dataset
x=data1.iloc[:,2:] #restante
y=data1.iloc[:,1] #diagnosis

y.value_counts()
```

## 1. HoldOut Validation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
mod_score1=model.score(x_test,y_test)
mod_score1
```

## 2. K Fold Cross Validation

```python
from sklearn.model_selection import KFold
import numpy as np
from sklearn.model_selection import cross_val_score

model=DecisionTreeClassifier()
kfold_validation=KFold(10)

mod_score3=cross_val_score(model,x,y,cv=kfold_validation)
print(mod_score3)

print(np.mean(mod_score3))
```

## 3. Leave One Out Cross Validation(LOOCV) - Demora bastante !!

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score

model=DecisionTreeClassifier()
leave_val=LeaveOneOut()
mod_score2=cross_val_score(model,x,y,cv=leave_val)
print(np.mean(mod_score2))
```

## 4. Stratified K-Fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold

sk_fold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
mod_score4=cross_val_score(model,x,y,cv=sk_fold)
print(np.mean(mod_score4))
print(mod_score4)
```

## 5. Repeated Random Test-Train Split

```python
from sklearn.model_selection import ShuffleSplit

model=DecisionTreeClassifier()
s_split=ShuffleSplit(n_splits=10,test_size=0.30)
mod_score5=cross_val_score(model,x,y,cv=s_split)
print(mod_score5)
print(np.mean(mod_score5))
```


## De acordo com esses testes em especifico, a melhor técnica é K Fold

Essas técnicas, quando implementadas adequadamente, ajudam a garantir que o modelo aprenda as características gerais dos dados, sem se ajustar demais a um conjunto específico
